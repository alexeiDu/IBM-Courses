```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker = yf.Ticker("TSLA")
```

```python
def make_graph(data, title="", x_label="", y_label="", markers=None):
  """
  Creates a line graph with optional title, labels, and markers.

  Args:
      data (pandas.Series): The data to be plotted (usually a time series).
      title (str, optional): The title of the graph. Defaults to "".
      x_label (str, optional): The label for the x-axis. Defaults to "".
      y_label (str, optional): The label for the y-axis. Defaults to "".
      markers (str, optional): The type of markers to use on data points (e.g., 'o' for circles). Defaults to None.
  """

  plt.plot(data.index, data, marker=markers)  # Plot with markers (optional)
  plt.xlabel("Date")
  plt.ylabel("Close Price (USD)")
  plt.title("Tesla Stock Price (Past Year)")
  plt.grid(True)  # Add gridlines
  plt.show()
```

```python
tesla_data = yf.download('TSLA', period='Max')
```

```python
make_graph(tesla_data, title="Tesla Stock Price (Past Year)")
```
